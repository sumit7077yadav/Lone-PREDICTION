```python
# first let handle the missing  value of data set
# for categorical_columns we use mode imputation
# for numerical value we use a mean imputation
```

```python
# for categorical values
categorical = ['Gender','Dependents','Self_Employed']
for cat  in categorical:
    data[cat].fillna(data[cat].mode()[0],inplace=True)

    
     # data[cat].fillna(data[cat].mode()[0], inplace=True)
     # data[cat].fillna(data[cat].mode()[0],inplace=True)

# for numerical value
numerical = ['LoanAmount','Loan_Amount_Term','Credit_History']
for num in numerical:
    data[num].fillna(data[num].mean(), inplace =True)
    
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn as sns
```

```python
# Univariate : Histograms for numeric columns
numeric  = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']
```

```python
# plot histograms
plt.figure(figsize =(16,5))
for i, col in enumerate(numeric, 1):
    plt.subplot(1 ,3, i)
    sns.histplot(data[col],kde =True)
    plt.title(f'Histogram of {col}')
plt.tight_layout()
plt.show()



# # Box plot
plt.figure(figsize =(15,5))
for i,  col in enumerate(numeric, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(data[col])
    plt.title(f'Box Plot of{col}')
                
plt.tight_layout()
plt.show()





#  # Bar Charts for categorical columns

categorical = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area']
plt.figure(figsize=(18, 10))

for i,col in enumerate(categorical,1):
    plt.subplot(2, 3, i)
    sns.countplot(x=col, data=data)
    plt.title(f'Bar Chart of {col}')

plt.tight_layout()
plt.show()

# Univariate Analysis: Pie Charts for categorical columns
plt.figure(figsize=(18, 10))
for i, col in enumerate(categorical, 1):
    plt.subplot(2, 3, i)
    data[col].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
    plt.title(f'Pie Chart of {col}')
    plt.ylabel('')

plt.tight_layout()
plt.show()

    
```

```python
# Bivariate Analysis: Scatter plots for numeric columns
```

```python

plt.figure(figsize=(15, 5))

plt.subplot(2, 3, 1)
sns.scatterplot(x='ApplicantIncome', y='LoanAmount', data=data)
plt.title('Applicant Income vs Loan Amount')

plt.subplot(1, 3, 2)
sns.scatterplot(x='CoapplicantIncome', y='LoanAmount', data=data)
plt.title('Coapplicant Income vs Loan Amount')

plt.subplot(1, 3, 3)
sns.scatterplot(x='ApplicantIncome', y='CoapplicantIncome', data=data)
plt.title('Applicant Income vs Coapplicant Income')

plt.tight_layout()
plt.show()


# Pair plot to visualize interactions between multiple numeric variables
sns.pairplot(data[numeric])
plt.show()   


plt.figure(figsize=(18, 8))
for i, col in enumerate(categorical[:3], 1):
    plt.subplot(1, 3, i)
    sns.boxplot(x=col, y='LoanAmount', data=data)
    plt.title(f'Loan Amount by {col}')

plt.tight_layout()
plt.show()

# Violin plots for categorical vs numeric variables
plt.figure(figsize=(18, 8))
for i, col in enumerate(categorical[3:], 1):
    plt.subplot(1, 3, i)
    sns.violinplot(x=col, y='LoanAmount', data=data)
    plt.title(f'Loan Amount by {col}')

plt.tight_layout()
plt.show()
```
